```python
# Create a random array of num bw o:10 and an array on ones size (3, 5)

random_array_1 = np.random.randint(10, size=(3, 5))
random_array_ones = np.ones((3,5))
```

```python
random_array_1, random_array_ones
```

```python
random_total = random_array_1 + random_array_ones
random_total
```

```python
random_another_ones = np.ones((5, 3))
random_another_ones
```

```python
random_add = random_another_ones + random_total.T
```

```python
random_add
```

```python
random_array_ones = np.ones((3, 5))
random_array_ones
```

```python
random_subtract = random_add - random_array_ones.T
random_subtract
```

```python
random_multiply = random_array_ones * random_subtract.T
random_multiply
```

```python
random_multiply ** 2
```

```python
np.square(random_multiply)
```

```python
np.mean(random_multiply)
```

```python
np.max(random_multiply), np.min(random_multiply)
```

```python
np.std(random_multiply)
```

```python
np.sqrt(7.25)
```

```python
np.var(random_multiply)
```
